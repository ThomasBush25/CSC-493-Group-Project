```python
 # Following libraries should be imported to run the experiment........
import numpy as np
import pandas as pd
```

```python
# Download and save the file in your directory.
# Read the file using pandas library.
df = pd.read_csv("Medicalpremium.csv")
df
```

```python
df = df.assign(Age=pd.cut(df['Age'], 
                               bins=[0, 10.5, 20.5, 30.5, 40.5, 50.5, 60.5, 70.5, 999], 
                               labels=['0->10', '11->20', '21->30', '31->40', '41->50', '51->60', '61->70', '70+']))
```

```python
df = df.assign(Diabetes=pd.cut(df['Diabetes'], 
                               bins=[-0.5, 0.5, 999], 
                               labels=['No', 'Yes']))
```

```python
df = df.assign(BloodPressureProblems=pd.cut(df['BloodPressureProblems'], 
                               bins=[-0.5, 0.5, 999], 
                               labels=['No', 'Yes']))
```

```python
df = df.assign(AnyTransplants=pd.cut(df['AnyTransplants'], 
                               bins=[-0.5, 0.5, 999], 
                               labels=['No', 'Yes']))
```

```python
df = df.assign(AnyChronicDiseases=pd.cut(df['AnyChronicDiseases'], 
                               bins=[-0.5, 0.5, 999], 
                               labels=['No', 'Yes']))
```

```python
df = df.assign(HistoryOfCancerInFamily=pd.cut(df['HistoryOfCancerInFamily'], 
                               bins=[-0.5, 0.5, 999], 
                               labels=['No', 'Yes']))
```

```python
df = df.assign(NumberOfMajorSurgeries=pd.cut(df['NumberOfMajorSurgeries'], 
                               bins=[-0.5, 1.1, 999], 
                               labels=['Low', 'High']))
```
